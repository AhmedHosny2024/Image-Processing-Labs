```python
img = io.imread('pout.tif')
noisy_img  = random_noise(img, mode='s&p',amount=0.05)
io.imshow(noisy_img)
io.show()
```

```python
'''
(1) Your own median filter algorithm
'''
outputPixelValue=np.zeros((img.shape[0],img.shape[1]))
windowWidth=3
windowHeight=3
edgax=math.trunc(windowWidth/2)
edgay=math.trunc(windowHeight/2)
for x in range(edgax,img.shape[1]-edgax):
     for y in range(edgay,img.shape[0]-edgay):
            allocateArray=img[y-edgay:y+2-edgay,x-edgax:x+2-edgax]
            outputPixelValue[y][x]=np.median(allocateArray)
            
io.imshow(outputPixelValue,cmap='gray')
io.show()
```

```python
'''
(2) skimage median filter
'''
Median=median(image=noisy_img)
io.imshow(Median,cmap='gray')
io.show()
```

```python
'''
(3) Gaussion Filters
'''
Gaussian_8=gaussian(image=noisy_img,sigma=8)
Gaussian_3=gaussian(image=noisy_img,sigma=3)
Gaussian_02=gaussian(image=noisy_img,sigma=.2)
show_images([Gaussian_8,Gaussian_3,Gaussian_02],['Sigma=8','Sigma=3','Sigma=0.2'])
```
